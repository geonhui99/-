```python
class oz:
    def __init__(self, name, age, level):
        self.name = name
        self.age = age
        self.level = level

coding = oz("김코", 30, "A")
print("이름", coding.name)
print("나이", coding.age)
print("코딩레벨", coding.level)
```

```python
class oz:
    def __init__(self, name, age, level):
        self.name = name
        self.age = age
        self.level = level

    def who(self):
        print(f"이름: {self.name}, 나이: {self.age}, 코딩레벨: {self.level}")

coding = oz("김코", 30, "A")
coding.who()
```
